```python
#tries to find the transition probability from given state to next by given action. Done by calculating the difference between the current
#value and the expected value , taking pre rotation errors into account
def find_transition_probabaility(err_prob,current_state, action_number, next_state):
    x,y,h = current_state
    x1,y1,h1 = next_state
    
    del_h = h1 - h
    if(del_h == 11):
        del_h = -1
    elif(del_h == -11):
        del_h == 1
    elif(del_h ==10 ):
        del_h = -2
    elif(del_h == -10):
        del_h == 2
    if action_number == 1:
        action = "no move"
    elif action_number == 2:
        action = "just_forward"
    elif action_number ==3 :
        action = "forward_left"
    elif action_number == 4:
        action = "forward_right"
    elif action_number == 5:
        action = "just_backward"
    elif action_number ==6:
        action = "backward_left"
    elif action_number == 7:
        action = "backward_right"
    if(action == "no move"):
        if(del_h == 0 and (x1-x) == 0 and (y1-y) == 0):
            return 1
        else:
            return 0
    
    if( action == "just_forward"):
    
        if del_h == 0:
            if find_same_direction_forward(current_state, next_state):
                return(1-2*err_prob)
            else:
                return 0
        elif del_h == -1:
            if h in [2,5,8,11]:
                if find_move_forward_left(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_forward(current_state, next_state):
                    return err_prob
                else:
                    return 0
        elif del_h == 1:
            if h in [1,4,7,10]:
                if find_move_forward_right(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_forward(current_state, next_state):
                    return err_prob
                else:
                    return 0    
    if(action == "forward_left"):
        if del_h == -1:
            if find_same_direction_forward(current_state, next_state):
                return(1-2*err_prob)
            else:
                return 0
        elif del_h == -2:
            if h in [2,5,8,11]:
                if find_move_forward_left(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_forward(current_state, next_state):
                    return err_prob
                else:
                    return 0
        elif del_h == 0:
            if h in [1,4,7,10]:
                if find_move_forward_right(current_state, next_state):
                                           
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_forward(current_state, next_state):
                    return err_prob
                else:
                    return 0 
    if( action == "forward_right"):
        if del_h == 1:
            if find_same_direction_forward(current_state, next_state):
                return(1-2*err_prob)
            else:
                return 0
        elif del_h == 0:
            if h in [2,5,8,11]:
                if find_move_forward_left(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_forward(current_state, next_state):
                    return err_prob
                else:
                    return 0
        elif del_h == 2:
            if h in [1,4,7,10]:
                if find_move_forward_right(current_state, next_state):
                                           
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_forward(current_state, next_state):
                    return err_prob
                else:
                    return 0 
    if(action == "just_backward"):
        if del_h == 0:
            if find_same_direction_backward(current_state, next_state):
                return(1-2*err_prob)
            else:
                return 0
        elif del_h == -1:
            if h in [2,5,8,11]:
                if find_move_backward_left(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_backward(current_state, next_state):
                    return err_prob
                else:
                    return 0
        elif del_h == 1:
            if h in [1,4,7,10]:
                if find_move_backward_right(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_backward(current_state, next_state):
                    return err_prob
                else:
                    return 0
    if(action == "backward_left"):
        if del_h == -1:
            if find_same_direction_backward(current_state, next_state):
                return(1-2*err_prob)
            else:
                return 0
        elif del_h == -2:
            if h in [2,5,8,11]:
                if find_move_backward_left(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_backward(current_state, next_state):
                    return err_prob
                else:
                    return 0
        elif del_h == 0:
            if h in [1,4,7,10]:
                if find_move_backward_right(current_state, next_state):
                                           
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_backward(current_state, next_state):
                    return err_prob
                else:
                    return 0
    if( action == "backward_right"):
        if del_h == 1:
            if find_same_direction_backward(current_state, next_state):
                return(1-2*err_prob)
            else:
                return 0
        elif del_h == 0:
            if h in [2,5,8,11]:
                if find_move_backward_left(current_state, next_state):
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_backward(current_state, next_state):
                    return err_prob
                else:
                    return 0
        elif del_h == 2:
            if h in [1,4,7,10]:
                if find_move_backward_right(current_state, next_state):
                                           
                    return err_prob
                else:
                    return 0
            else:
                if find_same_direction_backward(current_state, next_state):
                    return err_prob
                else:
                    return 0
    
    return 0
```

```python
#performs value iteration to find the best value of each state and then its corresponding action
def value_iteration(array,discount,err_prob):
    t1= time.time()
    value_array = np.zeros((6,6,12))
    delta = 1
    #temp1 = 0
    while True:
        delta = 0
        
        for state in state_space:
            x,y,h = state
            temp = value_array[x][y][h]
            action = get_new_action_value(state,value_array,0.9,err_prob)
            #print(action)
            value_array[x][y][h] = np.max(action)
            #print(value_array[x][y][h])
            #value[x][y][h] = temp1
            #print("one loop done")
            
            #print(value[x][y][h])
            
            delta = max(delta,abs(temp-value_array[x][y][h]))
        if(delta < 0.1):
            break
            #print(state,next_state,delta)
    #print(value_array)        
    policy = np.zeros((6,6,12))
    for state in state_space:
        x,y,h = state
        action_new = get_new_action_value(state,value_array,0.9,err_prob)
        policy[x][y][h] = np.argmax(action_new)+1
    t2 = time.time() 
    print("time", t2-t1)
    return(value_array,policy)
```

```python
def get_new_action_value(state,value,discount,err_prob):
    
    A = np.zeros(7)
    x,y,h=state
    for every_action in range(1,8):
        #next_state_space = get_possible_next_states(err_prob,state,every_action)
        #print(A)
        for next_state in state_space:
            x1,y1,h1 = next_state
            A[every_action-1] += find_transition_probabaility(err_prob,state, every_action, next_state) *( (get_reward(x,y) + (discount * value[x1][y1][h1])))
    #print(A)
    d = [num for num in A if num]    
    return d   
```

```python
import numpy as np
def get_possible_next_states(err_prob,current_state,action_val):
    x,y,h = current_state
    trans_prob = []
    trans_nzero = []
    for j in range(0,12):
        trans_prob.append((find_transition_probabaility(err_prob,current_state, action_val, (x,y,j)),(x,y,j)))
        trans_prob.append((find_transition_probabaility(err_prob,current_state, action_val, (x+1,y,j)),(x+1,y,j)))
        trans_prob.append((find_transition_probabaility(err_prob,current_state, action_val, (x,y+1,j)),(x,y+1,j)))
        trans_prob.append((find_transition_probabaility(err_prob,current_state, action_val, (x-1,y,j)),(x-1,y,j)))
        trans_prob.append((find_transition_probabaility(err_prob,current_state, action_val, (x,y-1,j)),(x,y-1,j)))
#     for state in state_space:
#         trans_prob.append((find_transition_probabaility(err_prob,current_state, action_val,state), state))
#     for i in range(len(trans_prob)):
#         if trans_prob[i][0] != 0:
#                 #.append(trans_prob[i])
    #print(trans_prob)
    #bound check
    to_remove = []
    to_add = []
    for i in range(len(trans_prob)):
        x1,y1,h1 = trans_prob[i][1]
        #print("in")
        #print(trans_prob[i][1])
        
        if x1<0  or x1 > 5 or y1 <0 or y1>5:
            #print("border")
            #print(trans_prob[i][1])
            temp = trans_prob[i][0]
#             #temp1 = trans_prob[i]
#             #print(temp,trans_prob[i])
#             trans_prob.remove(temp1)
#             trans_prob.append((temp,(x,y,h1)))
            to_remove.append(trans_prob[i])
            to_add.append((temp,(x,y,h1)))
    l3 = [x for x in trans_prob if x not in to_remove]
    l3.extend(to_add)
    #print(l3)    
    for i in range(len(l3)):
        if l3[i][0] != 0:
            trans_nzero.append(l3[i][1])
    return trans_nzero
```
